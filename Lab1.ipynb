```python
# Line Graph - Alcohol content by quality
alcohol_by_quality = Rwine.groupby('quality')['alcohol'].mean()

plt.figure(figsize=(10, 6))
plt.plot(alcohol_by_quality.index, alcohol_by_quality.values, 
         marker='o', color='blue', linestyle='--',linewidth=2,label= 'Alcohol by quantity')

plt.title('Average Alcohol Content by Wine Quality Rating')
plt.xlabel('Quality Rating')
plt.ylabel('Alcohol Content (% by volume)')
plt.savefig('Aty_AlcoholVsQualityLine.png')
plt.legend()
plt.show()
```

```python
# Pie Chart - pH value ranges
Rwine['pH_range'] = pd.cut(Rwine['pH'], bins=[2.7, 3.1, 3.4, 3.7, 4.0], 
                        labels=['2.7-3.1', '3.1-3.4', '3.4-3.7', '3.7-4.0'])
pH_counts = Rwine['pH_range'].value_counts()

plt.figure(figsize=(9, 9))
plt.pie(pH_counts, labels=pH_counts.index, autopct='%1.1f%%',
        colors=['orange','green','lightblue','yellow'],
        startangle=90)
plt.title('Distribution of Red Wine pH Ranges')
plt.savefig('Aty_WinePHDistributionPie.png')
plt.show()
```

```python
# Box Plot - Citric acid content by quality
plt.figure(figsize=(10, 7))
Rwine.boxplot(column='citric acid', by='quality', grid=False, 
           patch_artist=True,
           boxprops=dict(facecolor='lightgreen'),
           medianprops=dict(color='black'))

plt.title('Citric Acid Content Distribution by Quality', fontsize=14)
plt.suptitle('') 
plt.xlabel('Quality Rating')
plt.ylabel('Citric Acid Content (g/dm)')
plt.savefig('Aty_CitricAcidByQualityBox.png')
plt.show()
```
